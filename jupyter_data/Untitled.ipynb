```python
import pandas as pd
import glob, os
from datetime import datetime, timedelta
import pytz

timezone = pytz.timezone("Europe/Madrid")

col_names = [
    'Día', 
    'Hora', 
    'Peaje', 
    'Periodo', 
    'PVPC', 
    'TEU', 
    'TCU', 
    'PVPC PERD', 
    'Coeficiente pérdidas estandar', 
    'CP', 
    'OC', 
    'OS', 
    'OM', 
    'Cargo capacidad', 
    'Servicio interrumpibilidad', 
    'CCVh RCVtovph',
    'CCVh RFE',
    'CCVh RMRv',
    'CCVh rutinaria',
    'SAh',
    'Otros sistema',
    'CDSVh',
    'Coste banda',
    'Coste reserva',
    'Coste restricciones técnicas diario',
    'PMH',
    'Componente intradiario 1',
    'Mercado diario'
 ]
```

```python
data = None

for file in glob.glob("historico/*.xls"):
    new = pd.read_excel(file, sheet_name='Tabla de Datos', names=col_names, header=None, skiprows=5, usecols="A:AB").dropna(axis=0, how='all')
    new.index = pd.to_datetime(new['Día'].apply(lambda x: timezone.localize(x))+new['Hora'].apply(lambda x: timedelta(hours=x-1))).apply(lambda x: timezone.normalize(x))
    new.drop(['Día', 'Hora'], 1, inplace=True)
    new['Periodo']=pd.to_numeric(new['Periodo'], downcast='integer')
    data = new if data is None else pd.concat([data, new])
```
